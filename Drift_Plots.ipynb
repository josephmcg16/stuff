```python
#@title Import Libraries
import plotly.graph_objects as go
import pandas as pd
import numpy as np

from google.colab import files
```

```python
#@title Plot Function
def plot_drift(
    df, 
    title, 
    xaxis_title="Nominal Weight (kg)", 
    yaxis_title="Calculated Error (%)", 
    legend_title="Date of Test"
    ):

    dates = np.unique(df['Date of Test'].to_numpy())
    fig = go.Figure()

    for date in dates:
      filtered_df = df[df[legend_title]==date]
      fig.add_trace(
          go.Scatter(x=filtered_df[xaxis_title], y=filtered_df[yaxis_title], 
                    mode='lines+markers', 
                    name=pd.to_datetime(str(date)).strftime('%Y.%m.%d'))
      )

    fig.update_layout(
        title=title,
        xaxis_title=xaxis_title,
        yaxis_title=yaxis_title,
        legend_title=legend_title
    )
    return fig
```

```python
#@title NOT0459
file = files.upload()
filename = list(file.keys())[0]
df = pd.read_excel(filename)

title = filename.split('.')[0]

fig = plot_drift(df, title)
fig.show()
fig.write_html(f"{title}.html")
```
